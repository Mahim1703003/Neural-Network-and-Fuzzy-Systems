```python
# fitness function
p = 19
sum = 0
fitness_list = list()
for i in range(30):
    for j in range(20):
        sum = sum + df.iloc[i,j] * (2 ** p)
        p = p - 1
    fitness_list.append(sum)
    sum = 0
    p = 19
fitness_list
```

```python
df['fitness'] = fitness_list
df
```

```python
df = df.sort_values(by='fitness', ascending=True)
```

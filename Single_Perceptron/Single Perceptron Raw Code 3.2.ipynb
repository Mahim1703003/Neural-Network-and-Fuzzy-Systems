# Single Perceptron for n bit data where 70% is train data and 30% is test data


## Generating n bit of data

```python
n = int(input('Enter Number of bits : '))
count = 0
i = n
string = 'bit_'
total_number = 2 ** n
```

```python
value = list()
dictionary = dict()

while i >= 1 :
    key = string + str(i)
    d = 2 ** count

    while len(value) != total_number:
        for j in range(d):
            value.append(0)
        for j in range(d):
            value.append(1)
    
    dictionary[key] = value
    value = list()
    count = count + 1
    i = i - 1
```

```python
l = list(dictionary.items())
l
```

```python
reversed_dictionary = dict()
i = n-1
while i >= 0:
    reversed_dictionary[l[i][0]] = l[i][1]
    i = i - 1
reversed_dictionary
```

```python
dictionary = reversed_dictionary
dictionary
```

```python
output = dictionary['bit_1']
output
```

```python
import pandas as pd

df = pd.DataFrame(data=dictionary)
df
```

```python
df['Output'] = output
df
```

```python
df = df.drop('Output',axis=1)
df
```

## Generating n number of random weights ,  a fixed threshold value and a fixed learning rate

```python
n
```

```python
import numpy as np
```

```python
np.random.seed(42)
weights = np.random.rand(n)
weights
```

```python
threshold = 0.5
threshold
```

```python
learning_rate = 0.4
learning_rate
```

## Train Test

```python
train_percentage = 70
test_percentage = 100 - train_percentage

print('Train Percentage :',train_percentage)
print('Test Percentage :',test_percentage)
```

```python
import math

no_of_train_data = math.ceil(( total_number * train_percentage ) / 100)
no_of_test_data = total_number - no_of_train_data

print('No of Train Data :',no_of_train_data)
print('No of Test Data :',no_of_test_data)
```

## Adjusting Weights

```python
total_number
```

```python
# df.columns
# df.columns[0]
# df[df.columns[0]]
df[df.columns[0]]
```

```python
counter = 0

while counter!=no_of_train_data :

    for i in range(no_of_train_data):
        summation = 0
        
        for j in range(n):
            summation = summation + df[df.columns[j]][i] * weights[j]
        
        if summation >= threshold :
            predicted_output = 1

        else:
            predicted_output = 0

        if predicted_output == output[i]:
            counter = counter + 1
        
        else:
            difference = output[i] - predicted_output
            for j in range(n):
                weights[j] = weights[j] + learning_rate * difference * df[df.columns[j]][i]
            counter = 0
            break
```

```python
weights
```

```python
# proof that these weights are valid for train data

for i in range(no_of_train_data) :
    summation = 0

    for j in range(n):
        summation = summation + df[df.columns[j]][i] * weights[j]

    if summation >= threshold :
        predicted_output = 1
    else:
        predicted_output = 0
    
    print(predicted_output)
```

```python
right = 0
wrong = 0

for i in range(no_of_train_data,total_number) :
    summation = 0

    for j in range(n):
        summation = summation + df[df.columns[j]][i] * weights[j]

    if summation >= threshold :
        predicted_output = 1

    else:
        predicted_output = 0
    
    if predicted_output == output[i]:
        right = right + 1

    else:
        wrong = wrong + 1

accuracy = ( right * 100 ) / no_of_test_data

print('No of Test Data :',no_of_test_data)
print('Right :',right)
print('Wrong :',wrong)
print('Accuracy :',accuracy)
```
